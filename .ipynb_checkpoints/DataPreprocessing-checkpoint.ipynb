## Data Analysis-Numerical

```python
data=pd.read_csv('googleplaystore.csv').dropna()
```

## Q 1.Finding the average

```python
print('Average Rating of Apps:',float(str(sum(data['Rating']/len(data['Rating'])))[:4]))
```

## Q2.How many apps are there with rating 5

```python
c=0

for i in data['Rating']:
    if i==5.0:
        c+=1
print("No. of Rating 5 Apps:",c)
```

## Q3.How Many Range between 4 to 4.5

```python

c=0
for i in data['Rating']:
    if i>4 and i<4.5:
        c+=1
print("No of apps Rating between 4 to 4.5:",c)
```
